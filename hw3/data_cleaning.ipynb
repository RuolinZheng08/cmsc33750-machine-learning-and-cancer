# Clean up response

```python
ic50 = pd.read_csv('data/v17_fitted_dose_response.tsv', sep='\t')
```

```python
ic50['DRUG_ID'].unique().shape # 265 drugs
```

```python
ic50['COSMIC_ID'].unique().shape # 1074 cell lines
```

```python
# keep DRUG_ID, COSMIC_ID, LN_IC50
ic50_df = ic50[['DRUG_ID', 'COSMIC_ID', 'LN_IC50']].copy()
ic50_df.rename(columns={'DRUG_ID': 'drug_id', 'COSMIC_ID': 'cell_line_id',
                       'LN_IC50': 'log_IC50'}, inplace=True)
ic50_df.sort_values(by=['drug_id', 'cell_line_id'], inplace=True)
ic50_df.reset_index(drop=True, inplace=True)
```

```python
ic50_df.head()
```

```python
ic50_df.to_csv('data/IC50.csv', index=False)
```

# Inspect both DFs

```python
# a lot of unmatched cell lines between our two dfs
print(set(df['cell_line_id'].unique()) - set(ic50_df['cell_line_id'].unique()))
print(set(ic50_df['cell_line_id'].unique()) - set(df['cell_line_id'].unique()))
```

```python
# experiment with accessing data
cell_line_ids = ic50_df.loc[ic50_df['drug_id'] == 1, 'cell_line_id']
# 393 cell lines
print(ic50_df.loc[ic50_df['drug_id'] == 1, 'cell_line_id'].shape)
temp = df.loc[df['cell_line_id'].isin(cell_line_ids)]
print(temp.shape) # only 373 cell lines have gene expression data
print(ic50_df.loc[(ic50_df['drug_id'] == 1) &
                  (ic50_df['cell_line_id'].isin(temp['cell_line_id']))
                 ].shape)
temp.head()
```

```python
idx1 = ic50_df.loc[ic50_df['drug_id'] == 1, 'cell_line_id']
idx2 = df.loc[df['cell_line_id'].isin(idx1), 'cell_line_id']
print(idx1.shape, idx2.shape, 
      len(set(idx1).intersection(set(idx2))),
      len(set(idx1).intersection(set(df['cell_line_id'].unique()))))
```
